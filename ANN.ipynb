```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.layers import Dropout
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

# # Encoding categorical data
#### Label Encoding the "Gender" column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
print(X)
```

```python
# One Hot Encoding the "Geography" column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

# Part 2 - Building the ANN

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initializing the ANN
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(Dropout(rate=0.1))#drop out pour améliorer l'ann bieé et variance
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(Dropout(rate=0.1))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Predicting the Test set results
y_pred = ann.predict(X_test)
```

```python
print(y_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

# évaluer l'ann


```python
from keras.wrappers.scikit_learn import KerasClassifier

from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    ann = tf.keras.models.Sequential()
    ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
    ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
    ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
    ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return ann

```

```python
ann =KerasClassifier(build_fn=build_classifier, batch_size = 10, epochs = 100)
precisions=cross_val_score(estimator=ann,X=X_train,y=y_train,cv=10)#n_jobs=-1)#pour utiliser tt les couer
```

# Ajuster l'ann 

```python
from keras.wrappers.scikit_learn import KerasClassifier

from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(optimizer):
    ann = tf.keras.models.Sequential()
    ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
    ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
    ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
    ann.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return ann
```

```python
ann =KerasClassifier(build_fn=build_classifier)
parameters={"batch_size":[32,25],
            "epochs":[100],
            "optimizer":["adam","rmsprop"]
    
}
grid_search= GridSearchCV(estimator =ann,
                          param_grid = parameters,
                          scoring = "accuracy",
                          cv=5) 
```

```python
grid_search = grid_search.fit(X_train,y_train)
best_parameters =grid_search.best_params_
best_precision = grid_search.best_score_
```

```python
print(best_parameters)
print(best_precision)
```
